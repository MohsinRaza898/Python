### Task2:
Write a program that keep on accepting number from the user until user enters Zero. Display the sum and average of all the numbers.

Hint: Use while loop with flag.

```python
numbers= []
sum = 0
flag = True
while flag:
   number = int(input("enter numbers :"))
   if number==0:
       flag = False
   else:
    numbers.append(number)
    sum += number

if len(numbers)>0:
  length = len(numbers)
  average = sum / length
  print("The sum of given numbers is :",sum)
  print("The lenth of given numbers is :",length)
  print("The average of given numbers is :",average)
else:
  print("no number entered.")
```

### Task9:
Write a program that accepts a sentence and calculate the number of letters and digits.

Suppose the following input is supplied to the program:

hello world! 123

Then, the output should be:

LETTERS 10

DIGITS 3


```python
sentence = input("Enter a sentence")
letter_count = 0
digit_count  = 0

for char in sentence:
  if char.isalpha():
    letter_count  += 1
  elif char.isdigit():
    digit_count  += 1
print("letters",letter_count )
print("digits",digit_count )
```

### Task17:
You need to write a code which accepts a sequence of comma separated 4 digit binary numbers as its input and then check whether they are divisible by 5 or not.

The numbers that are divisible by 5 are to be printed in a comma separated sequence.

Example:

0100,0011,1010,1001

Then the output should be:
1010

Reference:
1) https://www.datacamp.com/.../python-data-type-conversion

2) https://stackoverflow.com/questions/21765779/converting-binary-to-decimal-integer-output?fbclid=IwAR3DD_oiq5pn99IjyxSJgojvB5W6M4JqQtv4EUl_IF_DtmuRXlP-DswcKAA

```python
binary_sequence = input("Enter a sequence of comma-separated 4-digit binary numbers: ").split(',')

divisible_by_5 = []

# Check divisibility and store numbers divisible by 5
for binary_num in binary_sequence:
    decimal_num = int(binary_num, 2)
    print(decimal_num)
    if decimal_num % 5 == 0:
        divisible_by_5.append(binary_num)

# Print the result
print(','.join(divisible_by_5))
```
