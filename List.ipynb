#### Task4: Sorting an array in ascending order
If user supplies an array(list) like arr = [5, 2, 8, 7, 1].


The output should be like:

Elements of original array:
5 2 8 7 1

Elements of array sorted in ascending order:
1 2 5 7 8

Note: We need to write our own sorting algorithm using sort function of list is not acceptable.

```python
arr = [5, 2, 8, 7, 1]

n = len(arr)
for i in range(n):
    for j in range(0, n-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]
            print(arr)
print(end='\n')
print(arr)
```

```python
arr = [5, 2, 8, 7, 1]
n = len(arr)
swapped = True

while swapped:
    swapped = False  # Reset the swapped flag
    for i in range(n - 1):
        if arr[i] > arr[i + 1]:
            arr[i], arr[i + 1] = arr[i + 1], arr[i]
            swapped = True
            print(arr)
print(end='\n\n')
print(arr)
```

#### Task6: Finding the max of the inputted array

```python
arr = [5, 2, 8, 7, 1]
max_value = arr[0]
for i in arr:
    if i > max_value:
      max_value = i
print("Max value in the array is :", max_value)
```
